```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   ( ,  BGR  )
img_bgr = cv2.imread('example.jpg')

# OpenCV   
img_gray_opencv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# NumPy     
blue_channel = img_bgr[:, :, 0]
green_channel = img_bgr[:, :, 1]
red_channel = img_bgr[:, :, 2]

#     (0.299*R + 0.587*G + 0.114*B)
img_gray_weighted = (0.114 * blue_channel + 0.587 * green_channel + 0.299 * red_channel).astype(np.uint8)

# NumPy      ((R + G + B) / 3)
img_gray_mean = (blue_channel/3.0 + green_channel/3.0 + red_channel/3.0).astype(np.uint8)

#   
plt.figure(figsize=(12, 8))

# 1.   
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))  # BGR RGB  
plt.title('Original Color Image')
plt.axis('off')

# 2. OpenCV Grayscale   
plt.subplot(2, 2, 2)
plt.imshow(img_gray_opencv, cmap='gray')
plt.title('OpenCV Grayscale')
plt.axis('off')

# 3. NumPy     
plt.subplot(2, 2, 3)
plt.imshow(img_gray_weighted, cmap='gray')
plt.title('NumPy Weighted Grayscale (0.299*R + 0.587*G + 0.114*B)')
plt.axis('off')

# 4. NumPy      
plt.subplot(2, 2, 4)
plt.imshow(img_gray_mean, cmap='gray')
plt.title('NumPy Mean Grayscale ((R + G + B) / 3)')
plt.axis('off')

#  
plt.tight_layout()
plt.show()
```
