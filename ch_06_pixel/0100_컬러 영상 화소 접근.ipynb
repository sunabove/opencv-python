```python
import cv2
import matplotlib.pyplot as plt

#  
img = cv2.imread('example.jpg')

#     ( )
pixel = img[100, 200]      # (100, 200)  BGR 
blue = img[100, 200, 0]    # Blue  
green = img[100, 200, 1]   # Green  
red = img[100, 200, 2]     # Red  

#     (: (100, 200)   )
img[100, 200] = [255, 0, 0]

#     
img[50:250, 100:400] = [0, 255, 0]

# OpenCV  BGR   RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# NumPy  BGR RGB 
img_rgb = img[:,:,::-1]  #     (BGR -> RGB)

#  
plt.imshow(img_rgb)
plt.title('Color image pixel access')
plt.axis('off')  #  
plt.show()
```
