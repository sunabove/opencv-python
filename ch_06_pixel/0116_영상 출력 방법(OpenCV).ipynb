```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   (OpenCV  BGR  )
img_bgr = cv2.imread('example.jpg')

# NumPy  BGR RGB 
img_rgb = img_bgr[:, :, ::-1]  #     (BGR -> RGB)
# OpenCV   BGR RGB 
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)  # BGR  RGB  
plt.title('Exmaple Image')
plt.axis('off')
plt.show()
```
