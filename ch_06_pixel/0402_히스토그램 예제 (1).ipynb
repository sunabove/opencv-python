```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("histogram_example.jpg", cv2.IMREAD_GRAYSCALE)

#  
# cv2.calcHist(images, channels, mask, histSize, ranges)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

rows, cols = 1, 2
chart_idx = 0
plt.figure(figsize=(cols*7, rows*3))

plt.subplot(rows, cols, chart_idx := chart_idx + 1 )
plt.title( 'Image' )
plt.imshow( img, cmap="gray" )
plt.axis( "on" )

#   Bar Chart  
plt.subplot(rows, cols, chart_idx := chart_idx + 1 )
plt.title('Histogram')
plt.bar(range(256), hist.ravel(), width=1, color='blue' )
plt.xlabel('Pixel Intensity')
plt.ylabel('Pixel Count')
plt.xlim([0, 255])  # x  0 255 
# x tick    (: 0, 25, 50, ..., 255)
tick_values = np.arange(0, 256, 25)
tick_texts = [ f"{tick}" for tick in tick_values ]
plt.xticks(tick_values, tick_texts, rotation=45)  #  tick  

plt.grid(axis='y', linestyle='--', alpha=0.7)

#  
plt.show()
```
