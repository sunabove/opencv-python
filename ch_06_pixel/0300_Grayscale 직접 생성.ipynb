```python
# grayscale 영상 생성
import numpy as np, cv2, matplotlib.pyplot as plt

image1 = np.zeros((50, 512), np.uint8) # 50 x 512 영상 생성
image2 = np.zeros((50, 512), np.uint8)

rows, cols = image1.shape[:2]

for i in range(rows):  # 행렬 전체 조회
    for j in range(cols):
        image1[i, j] = j//2        # 화소값 점진적 증가
        image2[i, j] = j//20*10    # 계단 현상 증가
    pass
pass

# 이미지들을 비교하여 출력
plt.figure(figsize=(12, 4))

# 1. 컬러 이미지 출력
plt.subplot(2, 1, 1)
plt.imshow( image1, cmap="gray")  # BGR을 RGB로 변환하여 출력
plt.title('Original Color Image')
plt.axis('off')

# 1. 컬러 이미지 출력
plt.subplot(2, 1, 2)
plt.imshow( image2, cmap="gray")  # BGR을 RGB로 변환하여 출력
plt.title('Original Color Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
