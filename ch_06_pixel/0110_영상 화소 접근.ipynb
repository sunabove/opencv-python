```python
import cv2
import matplotlib.pyplot as plt

#  
img = cv2.imread('example.jpg')

#     (: (100, 200)   )
img[100, 200] = [255, 0, 0]

#     
img[50:150, 100:200] = [0, 255, 0]

# OpenCV  BGR   RGB 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# NumPy  BGR RGB 
img_rgb = img[:,:,::-1]  #     (BGR -> RGB)

#  
plt.imshow(img_rgb)
plt.title('Modified Image')
plt.axis('off')  #  
plt.show()
```
