```python
import cv2, numpy as np, matplotlib.pyplot as plt

# 이미지 불러오기 (흑백 이미지로 변환)
image = cv2.imread('example.jpg', cv2.IMREAD_GRAYSCALE)

# 두 개의 다른 표준 편차를 가진 가우시안 블러를 적용
blurred_image_1 = cv2.GaussianBlur(image, (5, 5), 0)    # 표준편차가 작은 가우시안
blurred_image_2 = cv2.GaussianBlur(image, (15, 15), 0)  # 표준편차가 큰 가우시안

# 두 이미지를 빼서 DoG (Difference of Gaussian) 구하기
dog = blurred_image_1 - blurred_image_2

# 음수를 절댓값으로 변환하여 에지 강조
dog_abs = np.abs(dog).astype(np.uint8)

# 폰트 사이즈
fs = 30

# 결과 출력
plt.figure(figsize=(16, 16))

plt.subplot(2, 2, 1)
plt.title("Original Image", fontsize=fs)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("DoG Edge Detection", fontsize=fs)
plt.imshow(dog_abs, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
