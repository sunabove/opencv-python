```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1.  
image = cv2.imread('morphology_j.png', cv2.IMREAD_GRAYSCALE)

# 2.   (     )
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# 3.    (  5x5 )
kernel_square = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))  #  
kernel_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))  #  
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))  #  

# 4.      (  )
morph_open_square = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel_square)
morph_close_square = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel_square)
morph_gradient_square = cv2.morphologyEx(binary_image, cv2.MORPH_GRADIENT, kernel_square)
morph_tophat_square = cv2.morphologyEx(binary_image, cv2.MORPH_TOPHAT, kernel_square)
morph_blackhat_square = cv2.morphologyEx(binary_image, cv2.MORPH_BLACKHAT, kernel_square)

morph_open_ellipse = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel_ellipse)
morph_close_ellipse = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel_ellipse)
morph_gradient_ellipse = cv2.morphologyEx(binary_image, cv2.MORPH_GRADIENT, kernel_ellipse)
morph_tophat_ellipse = cv2.morphologyEx(binary_image, cv2.MORPH_TOPHAT, kernel_ellipse)
morph_blackhat_ellipse = cv2.morphologyEx(binary_image, cv2.MORPH_BLACKHAT, kernel_ellipse)

morph_open_cross = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel_cross)
morph_close_cross = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel_cross)
morph_gradient_cross = cv2.morphologyEx(binary_image, cv2.MORPH_GRADIENT, kernel_cross)
morph_tophat_cross = cv2.morphologyEx(binary_image, cv2.MORPH_TOPHAT, kernel_cross)
morph_blackhat_cross = cv2.morphologyEx(binary_image, cv2.MORPH_BLACKHAT, kernel_cross)

# 5.  
plt.figure(figsize=(20, 15))
fs = fontsize = 14

#    -     
plt.subplot(3, 5, 1)
plt.title("Opening (Square Kernel)", fontsize=fs)
plt.imshow(morph_open_square, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 2)
plt.title("Closing (Square Kernel)", fontsize=fs)
plt.imshow(morph_close_square, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 3)
plt.title("Gradient (Square Kernel)", fontsize=fs)
plt.imshow(morph_gradient_square, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 4)
plt.title("Tophat (Square Kernel)", fontsize=fs)
plt.imshow(morph_tophat_square, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 5)
plt.title("Blackhat (Square Kernel)", fontsize=fs)
plt.imshow(morph_blackhat_square, cmap='gray', interpolation='nearest')
plt.axis("off")

#    -     
plt.subplot(3, 5, 6)
plt.title("Opening (Ellipse Kernel)", fontsize=fs)
plt.imshow(morph_open_ellipse, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 7)
plt.title("Closing (Ellipse Kernel)", fontsize=fs)
plt.imshow(morph_close_ellipse, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 8)
plt.title("Gradient (Ellipse Kernel)", fontsize=fs)
plt.imshow(morph_gradient_ellipse, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 9)
plt.title("Tophat (Ellipse Kernel)", fontsize=fs)
plt.imshow(morph_tophat_ellipse, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 10)
plt.title("Blackhat (Ellipse Kernel)", fontsize=fs)
plt.imshow(morph_blackhat_ellipse, cmap='gray', interpolation='nearest')
plt.axis("off")

#    -     
plt.subplot(3, 5, 11)
plt.title("Opening (Cross Kernel)", fontsize=fs)
plt.imshow(morph_open_cross, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 12)
plt.title("Closing (Cross Kernel)", fontsize=fs)
plt.imshow(morph_close_cross, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 13)
plt.title("Gradient (Cross Kernel)", fontsize=fs)
plt.imshow(morph_gradient_cross, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 14)
plt.title("Tophat (Cross Kernel)", fontsize=fs)
plt.imshow(morph_tophat_cross, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.subplot(3, 5, 15)
plt.title("Blackhat (Cross Kernel)", fontsize=fs)
plt.imshow(morph_blackhat_cross, cmap='gray', interpolation='nearest')
plt.axis("off")

plt.tight_layout()
plt.show()
```
