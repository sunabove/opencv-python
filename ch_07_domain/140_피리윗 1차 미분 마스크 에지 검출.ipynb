```python
import cv2, numpy as np, matplotlib.pyplot as plt

#   (  )
image = cv2.imread('example.jpg', cv2.IMREAD_GRAYSCALE)

#   (x ) 
Gx = np.array([[-1, -1, -1],
               [0, 0, 0],
               [1, 1, 1]])

#   (y ) 
Gy = np.array([[-1, 0, 1],
               [-1, 0, 1],
               [-1, 0, 1]])

#     (convolution)
edge_x = cv2.filter2D(image, -1, Gx)  # x  
edge_y = cv2.filter2D(image, -1, Gy)  # y  

#       
edges_combined = cv2.addWeighted(edge_x, 0.5, edge_y, 0.5, 0)

#  
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title("Prewitt Edge X (Gx)")
plt.imshow(edge_x, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title("Prewitt Edge Y (Gy)")
plt.imshow(edge_y, cmap='gray')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title("Combined Prewitt Edges (Gx + Gy)")
plt.imshow(edges_combined, cmap='gray')
plt.axis('off')

plt.show()
```
