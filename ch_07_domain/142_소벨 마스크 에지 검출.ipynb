```python
import cv2, numpy as np, matplotlib.pyplot as plt

#   (  )
image = cv2.imread('example.jpg', cv2.IMREAD_GRAYSCALE)

#   (x )  
Gx = np.array([[-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1]])

#   (y )  
Gy = np.array([[-1, -2, -1],
               [0, 0, 0],
               [1, 2, 1]])

#     (convolution)
edge_x = cv2.filter2D(image, -1, Gx)  # x  
edge_y = cv2.filter2D(image, -1, Gy)  # y  

#       
edges_combined = cv2.addWeighted(edge_x, 0.5, edge_y, 0.5, 0)

#  
fs = 30

#  
plt.figure(figsize=(16, 16))

plt.subplot(2, 2, 1)
plt.title("Original Image", fontsize=fs)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Sobel Edge X (Gx)", fontsize=fs)
plt.imshow(edge_x, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Sobel Edge Y (Gy)", fontsize=fs)
plt.imshow(edge_y, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Combined Sobel Edges (Gx + Gy)", fontsize=fs)
plt.imshow(edges_combined, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
