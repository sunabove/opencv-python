```python
import cv2, numpy as np, matplotlib.pyplot as plt

# 이미지 불러오기
image = cv2.imread('example.jpg')

# OpenCV는 BGR 형식으로 이미지를 읽으므로, RGB로 변환
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 평균 필터 커널 생성
ks = kernel_size = 13 # 커널 사이즈
kernel = np.ones((ks, ks), np.float32)/ks**2

# filter2D를 사용하여 커널을 이미지에 적용 (평균 블러)
blurred_image = cv2.filter2D(image_rgb, -1, kernel)

# 원본 이미지와 평균 블러 적용된 이미지 비교 출력
plt.figure(figsize=(10, 5))

# 원본 이미지 출력
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# 평균 블러링된 이미지 출력
plt.subplot(1, 2, 2)
plt.imshow(blurred_image)
plt.title( f'{ks}x{ks} Average Blurred Image' )
plt.axis('off')

plt.tight_layout()
plt.show()
```
