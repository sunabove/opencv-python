```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

#   ()
image_path = 'example.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

#   ( )
kernel = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

# SciPy   
convoluted_image = convolve2d(image, kernel, mode='same', boundary='fill', fillvalue=0)

#   0~255   
convoluted_image = np.clip(convoluted_image, 0, 255)

# matplotlib  
plt.figure(figsize=(10, 5))

#   
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

#   
plt.subplot(1, 2, 2)
plt.imshow(convoluted_image, cmap='gray')
plt.title('SciPy Convoluted Image (Edge Detection)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
